```python
import pandas as pd
```

```python
df1 = pd.read_csv('applications(2.0).csv')
df2 = pd.read_csv('industries(2.0).csv')
df1
df2
```

```python
df1
```

```python
# 1. Прибирання дублікатів за applicant_id
df1 = df1.drop_duplicates(subset='applicant_id')
df1
```

```python
#2. Заповнення відсутніх значень у 'External Rating' нулями
df1.loc[:, 'External Rating'] = df1['External Rating'].fillna(0)
df1
```

```python
# 3.Заповнення пропущених значень у 'Education level' текстом "Середня"
df1.loc[:, 'Education level'] = df1['Education level'].fillna("Середня")
df1
```

```python
merged_df = pd.merge(df1, df2, how='left', on='Industry')
merged_df
```

```python
# Заповнення відсутніх значень у 'Amount'
merged_df['Amount'] = merged_df['Amount'].fillna(0)
merged_df
```

```python
# Створення колонки "Рейтинг" зі значенням 0
merged_df['Rating'] = 0
merged_df
```

```python
# Умови для обчислення рейтингу
merged_df['Rating'] += ((merged_df['Age'] >= 35) & (merged_df['Age'] <= 55)) * 20
merged_df['Rating'] += (~merged_df['Applied at'].str.contains('Sat|Sun', regex=True)) * 20
merged_df['Rating'] += (merged_df['Marital status'] == 'Married') * 20
merged_df['Rating'] += (merged_df['Location'].str.contains('Київ|Київська область', regex=True)) * 10
merged_df['Rating'] += merged_df['Score'].fillna(0)  # Score з таблиці industries.csv
merged_df['Rating'] += (merged_df['External Rating'] >= 7) * 20
merged_df['Rating'] -= (merged_df['External Rating'] <= 2) * 20
merged_df
```

```python
# Умови для обнулення рейтингу
merged_df.loc[(merged_df['Amount'] == 0) | (merged_df['External Rating'] == 0), 'Rating'] = 0
merged_df
```

```python
# Залишити лише заявки з рейтингом більше нуля
filtered_df = merged_df[merged_df['Rating'] > 0]
filtered_df
```

```python
# Перетворення дати у формат datetime і додавання тижня
filtered_df.loc[:, 'Applied at'] = pd.to_datetime(filtered_df['Applied at'], format="%d.%m.%Y %H:%M", errors='coerce')
filtered_df.loc[:, 'Week'] = filtered_df['Applied at'].dt.isocalendar().week
filtered_df
```

```python
weekly_avg_rating = filtered_df.groupby('Week')['Rating'].mean().reset_index()
weekly_avg_rating
```
