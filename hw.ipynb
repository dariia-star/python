```python
# 2. Знаходимо спільні колонки між опитувальником та даними
questions = set(schema['qname'].unique()) & set(df_hw.columns)
```

```python
# Видаляємо ряди з пропущеними значеннями у цих колонках
respondents_with_all_answers = df_hw.dropna(subset=questions).shape[0]
#Кількість респондентів, які відповіли на всі запитання опитувальника
print(f"Кількість респондентів, які відповіли на всі запитання опитувальника: {respondents_with_all_answers}")
```

```python
# 4.Кількість респондентів, які працюють віддалено 
remote_workers = df_hw[df_hw['RemoteWork'] == 'Remote'].shape[0]
print(remote_workers)
```

```python
#5 Який відсоток респондентів програмує на Python?
# Кількість респондентів, які програмують на Python
python_users = df_hw[df_hw['LanguageHaveWorkedWith'].str.contains('Python', na=False)].shape[0]
print(python_users)
```

```python
# 5 Який відсоток респондентів програмує на Python?
# Відсоток респондентів , які програмують на Python
percentage_python = (python_users / total_respondents) * 100
print(percentage_python)
```

```python
# Кількість респондентів, які  навчалися програмувати за допомогою онлайн курсів
online_learners = df_hw['LearnCodeOnline'].str.contains('Online Courses', na=False).sum()
print(online_learners)
```

```python
python_programmers = df_hw[df_hw['LanguageHaveWorkedWith'].str.contains('Python', na=False)]
python_programmers
```

```python
# 7. Серед респондентів що програмують на Python в групуванні по країнам, яка середня та медіанна сума компенсації (ConvertedCompYearly) в кожній країні?
python_devs = df_hw[df_hw['LanguageHaveWorkedWith'].str.contains('Python', na=False)]
compensation_stats = python_devs.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).reset_index()
print(compensation_stats)
```

```python
# 8. Які рівні освіти мають 5 респондентів з найбільшою компенсацією?
top_5_education = df_hw.nlargest(5, 'ConvertedCompYearly')[['EdLevel', 'ConvertedCompYearly']]
print(top_5_education)
```
